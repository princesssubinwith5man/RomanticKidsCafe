```python
img = cv2.imread(dirs)
plt.imshow(img)
```

```python
dirs = dirs + '/' + 'Image' + '/' + '1.mp4'
cap = cv2.VideoCapture(dirs)
if cap.isOpened():                 # 캡쳐 객체 초기화 확인
    while True:
        ret, img = cap.read()      # 다음 프레임 읽기      --- ②
        if ret:                     # 프레임 읽기 정상
            plt.imshow(img) # 화면에 표시  --- ③
            cv2.waitKey(25)            # 25ms 지연(40fps로 가정)   --- ④
        else:                       # 다음 프레임 읽을 수 없슴,
            break                   # 재생 완료
```
