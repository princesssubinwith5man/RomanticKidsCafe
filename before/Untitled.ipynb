```python
dirs = dirs + '/' + 'Image' + '/' + '1.mp4'
cap = cv2.VideoCapture(dirs)
if cap.isOpened():                 #    
    while True:
        ret, img = cap.read()      #         --- 
        if ret:                     #   
            plt.imshow(img) #    --- 
            cv2.waitKey(25)            # 25ms (40fps )   --- 
        else:                       #     ,
            break                   #  
```
